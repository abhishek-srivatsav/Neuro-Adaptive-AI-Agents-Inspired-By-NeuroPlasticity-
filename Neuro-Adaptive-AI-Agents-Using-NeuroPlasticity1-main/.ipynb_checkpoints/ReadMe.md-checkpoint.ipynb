# 🧠 Neuro-Adaptive Multi-Agent System

## 📌 Project Overview

The **Neuro-Adaptive Multi-Agent System** is an advanced AI framework inspired by **human neuroplasticity**.  
It simulates how humans adapt when one sensory capability is limited, by introducing **specialized AI agents** that compensate for missing modalities.

This system integrates:
- **Text**
- **Audio**
- **Image**

and dynamically routes reasoning through **Normal**, **Blind**, **Mute**, and **LessHearable** agents.

The project is designed for:
- Assistive AI systems
- Inclusive human–AI interaction
- Cognitive intelligence research
- Multimodal reasoning visualization

---

## 🧠 Neuroplasticity-Inspired Design

Just as the human brain adapts when one sense is impaired:
- **Blind Agent** compensates using audio + text
- **Mute Agent** compensates using image + text
- **LessHearable Agent** compensates using memory summarization
- **Normal Agent** uses all modalities

The agents **collaborate**, exchange feedback, and store compact memories to improve future reasoning.

---

## 🤖 Agents in the System

| Agent | Description |
|-----|------------|
| **NormalPersonAgent** | Full multimodal reasoning (image + audio + text) |
| **BlindPersonAgent** | Audio & text reasoning (no vision) |
| **MutePersonAgent** | Image & text reasoning (no audio) |
| **LessHearableAgent** | Memory & summarization agent enabling hearing compensation |

> ⚠️ LessHearableAgent runs internally and is not triggered directly by the user.

## 🔐 Required Configuration

### 1️⃣ Gemini API Key

Create a `.env` file in the project root:

```env
GEMINI_API_KEY=your_google_gemini_api_key
```

If not provided:
- Gemini reasoning is skipped
- System continues with fallback logic

---

### 2️⃣ Required Models

Place these files inside the `models/` directory:

```text
img_encoder_15d.pt
aud_encoder_15d.pt
txt_encoder_15d.pt
```

Download and place VOSK model:

```text
models/vosk-model-small-en-us-0.15/
```

## ▶️ Running the Project

### Option 1️⃣ Run Core System (Terminal Mode)

```bash
python main8.py
```

- Tests modalities
- Runs agent reasoning in terminal

### Option 2️⃣ Run Web Application (Recommended)

```bash
python app.py
```
